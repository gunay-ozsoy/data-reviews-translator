## Review Translator

ðŸ‘‰ `reviews` datasetâ€™ini load edin ve 1-yÄ±ldÄ±z ratingâ€™e sahip reviewâ€™lardan bir Ã¶rnek (sample) seÃ§in.

```python
# Verileri yÃ¼kle
from olist.data import Olist
data = Olist().get_data()
```

ðŸ‘€ 20 adet dÃ¼ÅŸÃ¼k puan alan yorumdan oluÅŸan bir Ã¶rneklem seÃ§in (rastgele) ve bunu bir listeye dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n.

```python
import pandas as pd

order_reviews = data["order_reviews"]

one_star = order_reviews.loc[
    (order_reviews["review_score"] == 1) &
    (order_reviews["review_comment_message"].notna()) &
    (order_reviews["review_comment_message"].str.strip().ne(""))
]

sample_pt = one_star["review_comment_message"].sample(20, random_state=42).tolist()
sample_pt[:3]
```

ðŸ—£ Bu gÃ¶reve baÅŸlamadan Ã¶nce Ã¶nceden yÃ¼klediÄŸiniz `google_translator` paketini kullanarak bu yorumlarÄ± Ã§evirin:

```python
from googletrans import Translator

translator = Translator()

sample_en = []
for t in sample_pt:
    res = await translator.translate(t, src="pt", dest="en")
    sample_en.append(res.text)

translated_df = pd.DataFrame({"pt": sample_pt, "en": sample_en})
translated_df["en"].head(20)
```
